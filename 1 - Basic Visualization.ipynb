```python
def lighten_color(color, amount=0.5):
    """
    Lightens the given color by multiplying (1-luminosity) by the given amount.
    Input can be matplotlib color string, hex string, or RGB tuple.

    Examples:
    >> lighten_color('g', 0.3)
    >> lighten_color('#F034A3', 0.6)
    >> lighten_color((.3,.55,.1), 0.5)
    """
    import matplotlib.colors as mc
    import colorsys
    try:
        c = mc.cnames[color]
    except:
        c = color
    c = colorsys.rgb_to_hls(*mc.to_rgb(c))
    return colorsys.hls_to_rgb(c[0], 1 - amount * (1 - c[1]), c[2])
```

```python
import matplotlib.ticker as ticker

fig, ax = plt.subplots(1, figsize=(6, 4),dpi = 100)
fig.suptitle("Age and Salary Plot")
# plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=10.0)

ax2 = ax.twinx()

# ax.plot(df_1['Name'],  df_1['Age'],color=lighten_color('r',amount=1.8),    label ="Ages"   )
ax.plot(df_1['Name'],  df_1['Age'],color=lighten_color('r',0.7),    label ="Ages"   )
ax2.plot(df_1['Name'], df_1['Salary'],color=lighten_color('b',1.1), label ="Salaries"  )


ax.set_ylabel('Age',       rotation = 45, labelpad=20, fontsize = 12) 
ax2.set_ylabel('Salary',   rotation = 0,  labelpad=40, )
ax.set_xlabel('Name List', rotation = 0)


plt.setp(ax.xaxis.get_majorticklabels(),  rotation=45)
plt.setp(ax.yaxis.get_majorticklabels(),  rotation=0 )
plt.setp(ax2.yaxis.get_majorticklabels(), rotation=0 )


ax2.yaxis.set_major_formatter(ticker.EngFormatter())

ax.legend( bbox_to_anchor=(1.20,+0.75), loc="lower right",  bbox_transform=fig.transFigure)
ax2.legend(bbox_to_anchor=(1.20,+0.82), loc="lower right",  bbox_transform=fig.transFigure)

plt.show()
```
